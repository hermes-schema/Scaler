```python
# Q1. Sequence [1-N] & [N-1]
# Problem Description: Write a program to print all Natural numbers from 1 to N and from N to 1 inclusive, where N is given as an argument.
# Note1: First print from 1 to N then in new line print again from N to 1
# Note2: You need not take input in this problem, you need to only implement the function provided.
# Note3: There is 1 line gap between two output lines and a space between numbers in same line.

def sequence(n):
    '''input: n = A single integer for the range where n is inclusive
       output: First Print from 1 to N and new line Print from N to 1'''
    
    # YOUR CODE GOES HERE
    for i in range(1,n+1):
        print(i, end=' ')
    print('\n')

    while n>0:
        print(n, end =' ')
        n -= 1

sequence(5)
```

```python
# Q5. Star pattern
# Write a function to print a pattern similar to the one shown in the sample using n 

def star(n):
    '''input: n = An Integer value defines number of rows
       output: print the star pattern'''
    # YOUR CODE GOES HERE
    for i in range(n):
        for j in range(i, n):
            if i == 0 or j == n -1 or j == i:
                print("*", end='')

            else:
                print(" ", end='')
        print()

star(8)
```

```python
# Q6. Perfect Number
# Problem Statement: Write a function to check whether a given number n as an input to the function is a perfect number or not. 
# If the given integer is a perfect number return 1 else return 0.

def perfect_number(n):
    '''input: n = An Integer value 
       output: return 1 if it is perfect number else 0'''
    # YOUR CODE GOES HERE
    sum=0
    for i in range(1,n):
        if n%i==0:
            sum+=i
    if sum==n:
        return 1
    else:
        return 0

perfect_number(6)
```

```python
# Assignment 2: Second Largest

class Solution:
   
    def solve(self, A):
        # This loop calculates the largest element in the list.
        max_elem = -1
        for x in A:
            if x > max_elem:
                max_elem = x
        
        # This loop calculates the second largest element in the list.   
        second_max = -1
        for x in A:
            if x > second_max and x != max_elem:
                second_max = x
        return second_max
```

```python
def solution(lst):
  sz = len(lst)
  if sz<=1:
    return "sorted"
    for rightIndex in range(1,sz):
      leftIndex = rightIndex-1
      leftValue = lst[leftIndex]
      rightValue = lst[rightIndex]
      if leftValue>rightValue:
        return "not sorted"
    return "sorted"
  
my_list = [-10,20,20,30]
print(solution(my_list))

my_list = [1,4,7,2,8,0]
print(solution(my_list))
```

```python
# Q4. Row Sum
# You are given a 2D integer matrix A, return a 1D integer array containing row-wise sums of original matrix.

ans = []
row_size = len(A)
col_size = len(A[0])

for i in range(0, row_size):
  row_sum = 0
  for j in range(0, col_size):
    row_sum+=A[i][j]
  ans.append(row_sum)
return ans
```

```python
# Q10. Perfect Number
# Problem Statement: Write a function to check whether a given number n as an input to the function is a perfect number or not. 
# If the given integer is a perfect number return 1 else return 0.

def perfect_number(n):
    '''input: n = An Integer value 
       output: return 1 if it is perfect number else 0'''
    # YOUR CODE GOES HERE
    sum=0
    for i in range(1,n):
        if n%i==0:
            sum+=i
    if sum==n:
        return 1
    else:
        return 0

perfect_number(8)
```
